```python
# Pair plot to see relationships between pollutants and AQI
sns.pairplot(df[['aqi', 'pm2_5', 'pm10', 'co', 'no2', 'o3', 'so2', 'nh3']])
plt.show()

# Correlation matrix
corr_matrix = df[['aqi', 'pm2_5', 'pm10', 'co', 'no', 'no2', 'o3', 'so2', 'nh3']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
